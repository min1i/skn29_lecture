## 2. Series 구조 이해
#### 2.1 Series의 정의
- 1차원 레이블이 있는 배열
- 동일한 타입의 데이터를 저장
- 인덱스(Index)와 값(Value)으로 구성
```
Index    Value
------   -----
0        10
1        20
2        30
3        40
```

#### 2.2 Series의 구성 요소
2.2.1 값(Values)
- 실제 데이터가 저장되는 부분
- NumPy 배열로 저장됨
- 정수, 실수, 문자열 등 다양한 타입 가능

2.2.2 인덱스(Index)
- 각 데이터를 식별하는 레이블
- 기본값: 0부터 시작하는 정수
- 사용자 정의 인덱스 설정 가능
- 문자열, 날짜 등 다양한 타입 가능

#### 2.3 Series의 특징
- 레이블 기반 인덱싱: 위치와 레이블 모두로 접근 가능
- 자동 정렬: 연산 시 인덱스 기준으로 자동 정렬
- 결측치 처리: NaN(Not a Number)으로 결측치 표현
- 벡터화 연산: 반복문 없이 전체 데이터 연산 가능

---

## 3. Series 생성 방법

#### 3.1 리스트로 Series 생성

````python
import pandas as pd

# 기본 생성 (자동 인덱스)
s = pd.Series([10, 20, 30, 40])

# 사용자 정의 인덱스
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
````

#### 3.2 딕셔너리로 Series 생성

````python
# 딕셔너리의 키가 인덱스가 됨
data = {'서울': 9720846, '부산': 3404423, '인천': 2947217}
s = pd.Series(data)
````

#### 3.3 스칼라 값으로 Series 생성

````python
# 인덱스를 지정하면 값이 반복됨
s = pd.Series(100, index=['a', 'b', 'c'])
````

#### 3.4 NumPy 배열로 Series 생성

````python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
s = pd.Series(arr)
````

---

## 4. Series 데이터 접근

#### 4.1 인덱스를 이용한 접근
위치 인덱싱 vs 레이블 인덱싱
--> 레이블 인덱싱은 'a'를 key값으로 인식해서 안됨

````python
# 위치 인덱싱 (정수)
s[0]      # 첫 번째 요소
s[1:3]    # 슬라이싱
````

````python
# 레이블 인덱싱
s['a']    # 레이블이 'a'인 요소 --> a가 없으니까 에러
s[['a', 'c']]  # 여러 레이블 선택
````

#### 4.2 Boolean 인덱싱

````python
# 조건을 만족하는 데이터 선택
s[s > 20]
s[(s > 10) & (s < 40)]
````

#### 4.3 주요 속성
- s.values: 값 배열 반환
- s.index: 인덱스 객체 반환
- s.dtype: 데이터 타입
- s.shape: 형태 (크기)
- s.size: 요소 개수

````python
print(s.values)
print(s.index)
print(s.dtype)
print(s.shape)
print(s.size)
````

## 5. Series 기본 연산

#### 5.1 벡터화 연산

````python
# 모든 요소에 동일한 연산 적용
s + 10
s * 2
s ** 2
````

#### 5.2 통계 함수

````python
s.mean()      # 평균
s.sum()       # 합계
s.max()       # 최댓값
s.min()       # 최솟값
s.std()       # 표준편차
s.describe()  # 기술통계 요약
````

````python
import numpy as np

np.random.seed(42)  # 재현성을 위해 시드 설정
data = np.random.randint(10,100,50)
s = pd.Series(data)
print(s.describe())
print(s.describe()['50%']) # 50% 위치의 값
````

#### 5.3 Series 간 연산

````python
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])

# 인덱스가 일치하는 요소끼리 연산
s1 + s2
````

---

## 6. 결측치 처리

#### 6.1 결측치 확인 (NaN)

````python
s.isnull()    # 결측치면 True
s.isna()      # 결측치면 True
s.notnull()   # 결측치가 아니면 True
````

````python
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([10, 20, 30], index=['a', 'b', 'cc'])
print(s1 + s2)  # 인덱스 'cc'는 s1에 없으므로 결과는 NaN

s = s1 + s2

# 결측치 (NaN) 여부 확인
s.isna()  # 결측치면 True

# 결측치 (NaN) 요소 선택
s[s.isna()].index
````

#### 6.2 결측치 처리

````python
s.dropna()           # 결측치 제거
s.fillna(0)          # 결측치를 0으로 채우기
s.fillna(s.mean())   # 평균값으로 채우기
````

## 8. pandas series 실습 예제

##### 예제 1: 학생 성적 관리
- 5명의 학생 성적을 Series로 생성
- 평균, 최고점, 최저점 계산
- 80점 이상 학생 필터링

````python
# [예제 1] 학생 성적 관리

import pandas as pd

data = {
    '김동글': 90, 
    '김네모': 75, 
    '김삼각형': 66, 
    '김하트': 89, 
    '김별': 97,
}

# 5명의 학생 성적을 Series로 생성
s = pd.Series(data)

# 평균, 최고점, 최저점 계산
print(f'평균: {s.mean()}\n최솟값: {s.min()}\n최댓값: {s.max()}')

# 평균, 최고점, 최저점인 점수를 Series로 생성
pd.Series([s.mean(), s.min(), s.max()], index=['평균', '최솟값', '최댓값'])

# 80점 이상인 학생 필터링
s[s >= 80]
````

##### 예제 2: 도시별 인구 데이터
- 주요 도시의 인구를 딕셔너리로 Series 생성
- 인구 순으로 정렬
- 특정 인구 범위 필터링

````python
# [예제 2] 도시별 인구 데이터

# 주요 도시의 인구를 딕셔너리로 Series 생성
population = pd.Series({
    '서울': 9720846,
    '부산': 3404423,
    '대전': 1471040,
    '인천': 2947217,
    '대구': 2427954
})

# 인구 순으로 정렬 (내림차순)
print("[인구 순으로 정렬 (내림차순)]")
print(population.sort_values(ascending=False))

# 특정 인구 범위 필터링
print("\n[인구가 300만 명 이상인 도시]")
print(population[population >= 3000000])
````

##### 예제 3: 결측치가 있는 데이터
- 결측치를 포함한 Series 생성
- 결측치 확인 및 처리
- 통계량 계산

````python
# [예제 3] 결측치가 있는 데이터

np.random.seed(42)
data = np.random.randn(20)

# 20개 중에 5개를 랜덤으로 뽑기
index = np.random.randint(0,20,5)
data[index] = np.nan  # 선택된 인덱스에 NaN 할당

# 1. 결측치를 포함한 Series 생성
s = pd.Series(data)

# 2. 결측치 확인 및 처리
s.isna() # 결측치면 True
s.notnull() # 결측치가 아니면 True
s[s.notnull()].index # 결측치가 아닌 요소의 인덱스

# 결측치가 아닌 데이터만 추출하기
s.notnull()

# 결측치를 평균이나 최대 또는 최소, 중위수로 채우기



# 3. 통계량 계산 
#s.describe()
````
