# 3. 데이터 로드 및 저장

## 1. 데이터 입출력 개요

#### 1.1 Pandas가 지원하는 파일 형식
- **CSV**: 쉼표로 구분된 텍스트 파일
- **Excel**: .xlsx, .xls 파일
- **JSON**: JavaScript Object Notation
- **HTML**: 웹 페이지의 표
- **SQL**: 데이터베이스
- **Pickle**: Python 객체 직렬화
- **Parquet**: 컬럼 기반 저장 형식

#### 1.2 주요 입출력 메서드
```
읽기                     쓰기
-----------------       -----------------
pd.read_csv()          df.to_csv()
pd.read_excel()        df.to_excel()
pd.read_json()         df.to_json()
pd.read_html()         df.to_html()
pd.read_sql()          df.to_sql()
```

## 2. CSV 파일 다루기

#### 2.1 CSV 파일 읽기

- 기본 읽기

````python
import pandas as pd

# df = pd.read_csv('20220207.csv', encoding='cp949', header=None)
````

- 주요 파라미터

````python
df = pd.read_csv(
    '20250731.csv',
    sep=',',                    # 구분자 (기본값: 쉼표)
    header=0,                   # 헤더 행 번호 (기본값: 0)
    index_col=0,                # 인덱스로 사용할 열
    # names=['A', 'B', 'C'],      # 컬럼명 지정
    # usecols=['A', 'B'],         # 특정 컬럼만 읽기
    # dtype={'A': int, 'B': str}, # 데이터 타입 지정
    encoding='utf-8',           # 인코딩 (한글: 'cp949', 'utf-8')
    na_values=['NA', 'null'],   # 결측치로 인식할 값
    skiprows=2,                 # 처음 n개 행 건너뛰기
    nrows=100                   # n개 행만 읽기
)
````

- 인코딩 문제 해결

````python
# Windows에서 한글 깨짐 시
df = pd.read_csv('20220207.csv', encoding='cp949')

# UTF-8 BOM 문제
df = pd.read_csv('20220207.csv', encoding='utf-8-sig')
````
