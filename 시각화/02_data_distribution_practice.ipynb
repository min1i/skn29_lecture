# 데이터 분포

- 히스토그램
- kde (밀도)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 한글 폰트 설정 (Windows)
plt.rcParams['font.family'] = 'Malgun Gothic'

# 음수 부호가 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False  
```

```python
# 데이터 로드 mpg

mpg = sns.load_dataset('mpg')
print(f'전체 데이터 : {len(mpg)}')
mpg.describe()
```

```python
# 연비는 정규분포를 따를까?
# histogram

fig, ax = plt.subplots(figsize=(10,6))

n, vins, patches = ax.hist(mpg['mpg'], bins=25, color='skyblue', edgecolor='black', alpha=0.7, linewidth=0.2)

# 평균선 추가
mean_mpg = mpg['mpg'].mean()
ax.axvline(mean_mpg, color='red', linestyle='--', linewidth=2.5, label=f'평균: {mean_mpg:.1f} mpg')
ax.legend()

# 중앙선 추가
median_mpg = mpg['mpg'].median()
ax.axvline(median_mpg, color='blue', linestyle='-.', linewidth=2.5, label=f'중앙값: {median_mpg:.1f} mpg')
ax.legend()
```

```python
# 밀도 : 부드러운 분포 곡선
# 국가별 연비 분포

colors = {'usa': 'red', 'japan': 'blue', 'europe': 'green'}

fig, ax = plt.subplots(figsize=(12,6))
for origin, color in colors.items():
    data = mpg['origin'] == origin
    data = mpg[data]['mpg']
    data.plot.kde(ax=ax,color=color,linewidth=3, label=f'{origin}평균:{data.mean():.1f}')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# boxplot : 이상치, 사분위수

# 실린더 수별로 연비편차가 얼마나 다를까?
fig, ax = plt.subplots(figsize=(10,6))
mpg_clean =  mpg[mpg['cylinders'].isin([4,6,8])].copy()

box_data = [  mpg_clean[mpg_clean['cylinders'] == cyl]['mpg']  for cyl in [4,6,8]]
bp = ax.boxplot(box_data,label=['4기통','6기통','8기통'],patch_artist=True,notch=True,widths=0.6)

# box색상
colors = ['#F58727','#76F527','#27D3F5']
for patch,color in zip(bp['boxes'],colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.legend()

# 중앙값선 강조
for median in bp['medians']:
    median.set(color='black', linewidth=2)

plt.tight_layout()
plt.show()

# 통계요약
for cyl in [4,6,8]:
    data = mpg_clean[mpg_clean['cylinders'] == cyl]['mpg']
    q1,q3 = data.quantile([0.25,0.75])
    iqr = q3-q1
    print(f'{cyl}기통 - 중앙값:{data.median():.1f} IQR:{iqr:.1f} \
           이상치개수 : { len(data[(data > q3+1.5*iqr) | (data < q1 - 1.5*iqr)] ) }' )
```

```python
# violineplot 분포형태
# 연식에 따라 연비가분포가 어떻게 진화했는지?

fig, ax = plt.subplots(figsize=(12,6))
sns.violinplot(data=mpg, x='model_year', y='mpg',ax=ax,inner='quartitle')
plt.tight_layout()
plt.show()
```

```python
# histogram + KDE + boxplot
# 차량 무게 분포
fig,ax = plt.subplots(2,1,figsize=(12,10))

# 상단 histogram +   KDE
ax[0].hist(mpg['weight'], bins=30,density=True, label='histogram')
mpg['weight'].plot.kde(ax = ax[0],color='red',label='kde')

#하단 boxplot(제조 국가별)
box_data_weight = [mpg[mpg['origin'] == 0]['weight'] for o in ['usa','japan','europe']]
bp = ax[1].boxplot(box_data_weight,label=['usa','japan','europe'],
                   patch_artist=True,notch=True,widths=0.6)
for patch,color in zip(bp['boxes'],['red','green','blue']):
    patch.set_facecolor(color)
plt.tight_layout()
plt.show()    
```
