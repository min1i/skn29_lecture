```python
# 간단한 코퍼스를 활용 Word2Vec 학습, 단어간 유사도 확인

from gensim.models import Word2Vec
```

```python
# 토큰화된 문장
sentences = [
    ['I', 'love', 'natural', 'language', 'processing'],
    ['natural', 'language', 'processing', 'is', 'fun'],
    ['I', 'enjoy', 'learning', 'machine', 'learning']
]

# Word2Vec 모델 학습 (CROW 기반)
w2v_model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, epochs = 1000)
print(f'language 벡터 : {w2v_model.wv["language"]}')
print(f'natural, language 유사도: {w2v_model.wv.similarity("natural", "language")}')
print(f'processing, machine 유사도: {w2v_model.wv.similarity("processing", "language")}')
```

```python
# 글로벌 동시 등장확률

import gensim.downloader as api

print('Glove 모델 다운로드 중... 대략 66MB')
glove_model = api.load('glove-wiki-gigaword-50')
print('로딩완료\n')

print(f'king, queen 유사도 : {glove_model.similarity("king", "queen")}')
print(f'man, woman 유사도 : {glove_model.similarity("man", "woman")}')
```

```python
# 벡터 연산
result = glove_model.most_similar(positive=['king','woman'], negative=['man'], topn=3)
print(f'king-man-woman에 가장 가까운 단어 : {result}')
```

```python
# FastText (00V 문제 해결)
# 훈련 데이터에 없거나 오탈자에 대해 유연하게 동작하는지 확인

from gensim.models import FastText

# 실습 데이터가 작아서 과적합을 방지하기 위해 10000번 학습
ft_model = FastText(sentences, vector_size = 10, window = 2, min_count = 1, min_n=3, max_n=3, epochs = 10000)

# 학습 데이터에 존재하는 단어
print(f"language 벡터: {ft_model.wv['languate']}")

print(f"학습되지 않은 단어 : languag : {ft_model.wv['languag']}")
print(f"langug, languge의 유사도 : {ft_model.wv.similarity('language', 'languag')}")
```
